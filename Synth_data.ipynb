# Synth Data

```python
names=['A'+str(i) for i in range(100)]

tags=['blues', 'rock', 'jazz', 'techno', 'pop', 'dance',
      'yoga', 'trip', 'cinema', 'gym', 'mountain', 'beach', 
      'ski', 'bike', 'skate', 'swiming', 'basket', 'running']
```

```python
data=np.random.randint(0, 6, (len(names), len(tags)))
```

```python
s_data=pd.DataFrame(data, columns=tags, index=names)
s_data.head()
```

```python
squareform(pdist(s_data, 'euclidean'))
```

```python
similar=pd.DataFrame(1/(1 + squareform(pdist(s_data, 'euclidean'))), 
                         index=s_data.index, columns=s_data.index)

similar.head()
```

# New user

```python
n_rating=np.random.randint(0, 6, (1, len(tags)))
B0={k:v for k, v in list(zip(tags, n_rating[0]))}
B0
```

```python
s_data=s_data.append(B0, ignore_index=True)
names.append('B0')
s_data.index=names
```

```python
s_data.head()
```

```python
similar = pd.DataFrame(1/(1 + squareform(pdist(s_data, 'euclidean'))), 
                         index=s_data.index, columns=s_data.index)

similarities = similar['B0'].sort_values(ascending=False)[1:8]
similarities
```
